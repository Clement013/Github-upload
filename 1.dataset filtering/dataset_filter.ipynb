```python
voice_data = load_dataset("clt013/malay-speech-3k-rows-dataset_v2", token=True)
voice_data
```

```python
def play_audio(dataset, index):
    audio_sample = dataset[index]['audio']
    samples = audio_sample['array']
    sampling_rate = audio_sample['sampling_rate']
    duration = len(samples) / sampling_rate
    print("Duration:", duration, 'seconds')
    return Audio(data=samples, rate=sampling_rate)

play_audio(voice_data['train'], 2)
voice_data['train'][2]
```

```python
def play_audio_df(df, index):
    audio_sample = df['audio'][index]
    samples = audio_sample['array']
    sampling_rate = audio_sample['sampling_rate']
    duration = len(samples) / sampling_rate
    print("Duration:", duration, 'seconds')
    return Audio(data=samples, rate=sampling_rate)
```

```python
# create a paging loop to display the audio and text

page = 32
dataset_validate = df_test
for idx in range((page - 1) * 10, page * 10):
    if idx >= len(dataset_validate):
        break
    print("No:",idx)
    display(play_audio_df(dataset_validate, idx))
    display(dataset_validate['sentence'][idx])
```
